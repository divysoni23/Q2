```python
import pandas as pd

# Load the data
data = pd.read_csv('ToyotaCorolla.csv', encoding='ISO-8859-1')

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
import matplotlib.pyplot as plt

# Filter data by fuel type
diesel = data[data['Fuel_Type'] == 'Diesel']
cng = data[data['Fuel_Type'] == 'CNG']
petrol = data[data['Fuel_Type'] == 'Petrol']

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(diesel['Age_08_04'], diesel['Price'], color='blue', label='Diesel', alpha=0.6)
plt.scatter(cng['Age_08_04'], cng['Price'], color='green', label='CNG', alpha=0.6)
plt.scatter(petrol['Age_08_04'], petrol['Price'], color='red', label='Petrol', alpha=0.6)

# Add labels and title
plt.xlabel('Vehicle Age (in months)')
plt.ylabel('Price (in Euros)')
plt.title('Price Trends of Used Toyota Corolla Vehicles by Age and Fuel Type')
plt.legend(title='Fuel Type')
plt.grid(True)
plt.show()
```
